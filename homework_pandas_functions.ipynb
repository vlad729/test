```python
#1

import pandas as pd

ratings = pd.read_csv('ratings.csv')



def film_class(param):
    if param <= 2:
        class_ = 'low'
        return class_
    elif 2 < param <= 4:
        class_ = 'average'
        return class_
    else:
        class_ = 'high'
        return class_
        

ratings['film_class'] = ratings['rating'].apply(film_class)

ratings.head()
```
