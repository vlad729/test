```python
#1

import pandas as pd

movies = pd.read_csv('movies.csv')

ratings = pd.read_csv('ratings.csv')

rating_data = ratings.query('rating == 5.0').groupby(by = 'movieId').count().sort_values(by='rating', ascending=False)

result = rating_data.join(movies, on='movieId')

result
```
